```python
# Import module
from plot_xgt3 import quick_plot
import numpy as np
```

```python
data_dir='../data'
varname='T2'
levels=np.arange(240,300,5)
cmap='bwr'
year=1981
month=1
title=f'2m Temperature, {year}-{month}'
quick_plot(f'{data_dir}/{varname}', 
           levels=levels, cmap=cmap,
           year=year, month=month,
           title=title)
```
